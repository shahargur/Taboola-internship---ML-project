### Turn the features stored in arrays into vectors - categort_ids

```python
import numpy as np
import pandas as pd
data= pd.read_parquet('data_for_starship_2022.snappy.parquet', engine='pyarrow',columns=[ 'user_iab_category_ids', 'user_iab_category_clicks','user_iab_category_recs' ] )
```

```python
category_ids_arrys = data['user_iab_category_ids']
category_ids_arrys=category_ids_arrys.to_numpy()
all_items_id=np.concatenate(category_ids_arrys, axis=None)
category_ids = pd.DataFrame(all_items_id).value_counts().index.to_list()
index_list = []
for cell in category_ids:
    index_list.append(cell[0])
category_ctr_data = pd.DataFrame(0.0, index=np.arange(len(data)), columns=index_list)
category_ctr_data.shape
```

```python
category_ids_arrys = data['user_iab_category_ids']
category_clicks_arrys = data['user_iab_category_clicks']
category_recs_arrys = data['user_iab_category_recs']

for row_index in range(len(category_ids_arrys)):
    id_vals = category_ids_arrys[row_index]
    for i in range(len(id_vals)):
        if float(category_recs_arrys[row_index][i]) == 0:
            category_ctr_data.at[row_index, id_vals[i]] = 0
        else:
            category_ctr_data.at[row_index, id_vals[i]] = float(category_clicks_arrys[row_index][i])/float(category_recs_arrys[row_index][i]) 
category_ctr_data = category_ctr_data.add_suffix('_c_ctr')
```

```python
data.to_parquet('category_ctr_data.snappy.parquet',
              compression='snappy',engine='pyarrow')
```

### Turn the features stored in arrays into vectors - user_item

```python
items_ids_arrys = data['user_item_ids'].to_numpy()
all_items_id=np.concatenate(items_ids_arrys, axis=None)
vals = pd.DataFrame(all_items_id).value_counts().index.to_list()
count_vals= pd.DataFrame(all_items_id).value_counts()
```

```python
index_list = []
val_list = []
for cell in vals:
    index_list.append(cell[0])
for val_cell in count_vals:
    val_list.append(val_cell)
```

```python
# select the 0.01% most commen items
insertes_columns=index_list[:int(734910/10000)]
```

```python
items_clicks_data = pd.DataFrame(0.0, index=np.arange(len(data)), columns=insertes_columns)
```

```python
item_columms_set= set(insertes_columns)
items_ids_arrys = data['user_item_ids']
items_clicks_arrys = data['user_item_clicks']
for row_index in range(len(items_ids_arrys)):
    val = items_ids_arrys[row_index]
    for i in range(len(val)):
        if val[i] in item_columms_set:
            items_clicks_data.at[row_index, val[i]] = float(items_clicks_arrys[row_index][i])   
items_clicks_data = items_clicks_data.add_suffix('_i_clicks')
```

### Multi-hot encodinf to the feature target_named_entities_ids

```python
user_item_ids=np.concatenate(data['target_named_entities_ids'].to_numpy(), axis=None)
entities = pd.DataFrame(user_item_ids).value_counts().index.to_list()
count_vals= pd.DataFrame(user_item_ids).value_counts()
```

```python
entities_list = []
value_count_list = []
for cell in entities:
    entities_list.append(cell[0])
for val_cell in count_vals:
    value_count_list.append(val_cell)
```

```python
# select the 0.2% most common entities
insertes_columns=entities_list[:int(36369/500)]
```

```python
target_named_entities_ids_data = pd.DataFrame(0, index=np.arange(len(data)), columns=insertes_columns)
```

```python
columms_set= set(insertes_columns)
entities_arrys = data['target_named_entities_ids']
for row_index in range(len(entities_arrys)):
    val = entities_arrys[row_index]
    for i in range(len(val)):
        if val[i] in columms_set:
            target_named_entities_ids_data.at[row_index, val[i]] = 1
target_named_entities_ids_data = target_named_entities_ids_data.add_suffix('_entities_ids')
```

### clean noise

```python
data.target_item_age_in_hours = np.where(data.target_item_age_in_hours < 0, 0, data.target_item_age_in_hours)
```

### Turn the categoricals features into one-hot encoding

```python
categorical_features= ['quality_level_string','browser_platform_string', 'iab_category', 'title_detected_language', 'target_item_upper_taxonomy']
for c in categorical_features:
    df1= pd.get_dummies(data[c] , prefix=c)
    data.drop([c], axis=1, inplace=True)
    data = pd.concat([data, df1], axis=1)
```

### remove features that transformed and combine the new ones

```python
data.drop(['user_item_ids', 'user_item_clicks', 'user_syndicator_ids', 'user_syndicator_clicks', 'user_syndicator_recs', 
          'user_iab_category_ids', 'user_iab_category_clicks', 'user_iab_category_recs','target_named_entities_ids'], axis=1, inplace=True)
```

```python
data = pd.concat([data, target_named_entities_ids_data, items_clicks_data, category_ctr_data], axis=1)
```

```python
data.to_parquet('data_ready_for_model.snappy.parquet',
              compression='snappy',engine='pyarrow')
```

### some cleaning to make the data smaller

```python
counter = 0
to_remove= []
under_1000_of_1 = []
in_10_prcente =[]
almost_all = []
for column in data.columns:
    cv = pd.DataFrame(data[{column}]).value_counts()
   
    if cv.shape[0] == 2 and column != 'is_click':
        if cv.iloc[0] > 1000000 and cv.iloc[0] <1500000:
#             print(cv)
            counter += 1
            to_remove.append(column)
        if cv.iloc[1] <2500:
#             print(cv)
            counter += 1
            to_remove.append(column)
    elif cv.iloc[0] >2497500:
#         print(cv)
        to_remove.append(column)
        counter += 1
    elif cv.shape[0] == 2500000:
#         print(cv)
        to_remove.append(column)
        counter += 1

data.drop(to_remove, axis=1, inplace=True)
    
```

```python
data.drop(['syndicator_id','country_code', 'user_id', 'publisher_id', 'region', 'target_item_type_string', 
          'session_external_referrer', 'zipcode', 'city'], axis=1, inplace=True)
```
